```python
import pandas as pd
import numpy as np
```

```python
grid_crop = pd.read_csv('Grided_Harvest_30mins_2d.csv')
```

```python
aez_map = pd.read_csv('GTAP_AEZ_Mapping.csv')
```

```python
aez_map =  aez_map[['Longitude', 'Latitude', 'AEZ', 'CTRYNAME', 'CTRYCODE','GTAPCODE']]
```

```python
result_normal = pd.read_csv('harvst_aez_reg_all.csv')
result_extreme = pd.read_csv('harvst_aez_reg_extreme.csv')
```

```python
## Merge grid crop and aez_map
grid_crop.head()
```

```python
aez_map.head()
```

```python
grid_crop_aez = pd.merge(grid_crop,aez_map, on = ['Longitude','Latitude'],how = 'right')
```

```python
grid_crop_aez[(grid_crop_aez['CTRYNAME']=='Usa') & (grid_crop_aez['Soybeans']>100)]
```

```python
grid_crop_aez.shape
```

```python
grid_crop_aez_index = grid_crop_aez[['Longitude', 'Latitude','AEZ', 'CTRYNAME', 'CTRYCODE', 'GTAPCODE']]
```

```python
grid_crop_aez_index.shape
```

```python
### Merge with normal cases
# % changes
grid_crop_normal = pd.merge(grid_crop_aez_index,result_normal, on = ['AEZ','GTAPCODE'], how = 'left')
```

```python
grid_crop_normal.shape
```

```python
# % changes
grid_crop_extreme = pd.merge(grid_crop_aez_index,result_extreme, on = ['AEZ','GTAPCODE'], how = 'left')
```

```python
# Absolute changes
crop_names = ['Paddy_Rice', 'Wheat', 'Sorghum', 'Oth_CrGr',
       'Soybeans', 'palmf', 'Rapeseed', 'Oth_Oilseeds', 'Sugar_Crop',
       'OthAgri', 'Pasturecrop', ]
```

```python
grid_crop_normal_delta = grid_crop_aez_index.copy();    
grid_crop_extreme_delta = grid_crop_aez_index.copy();    
```

```python
for crop in crop_names:
    print(crop)
    grid_crop_normal_delta[crop] =  grid_crop_aez[crop] * grid_crop_normal[crop]/100
    grid_crop_extreme_delta[crop] = grid_crop_extreme[crop] * grid_crop_aez[crop]/100
```

```python
grid_crop_normal_delta_TArea = grid_crop_aez_index.copy();  
```

```python
grid_crop_normal_delta_TArea['CropH_Area'] = np.sum(grid_crop_normal_delta.iloc[:,6:16],axis = 1)
```

```python
grid_crop_normal_delta_TArea[grid_crop_normal_delta_TArea['CropH_Area']<0].GTAPCODE.unique()
```

```python
grid_crop_normal_delta_TArea.to_csv('TotalHarvestedAreaChanges_Gridded_Normal.csv',index=False)
```

```python
grid_crop_extreme_delta_TArea = grid_crop_aez_index.copy();  
grid_crop_extreme_delta_TArea['CropH_Area'] = np.sum(grid_crop_extreme_delta.iloc[:,6:16],axis = 1)
```

```python
grid_crop_extreme_delta_TArea.to_csv('TotalHarvestedAreaChanges_Gridded_Extreme.csv',index=False)
```
